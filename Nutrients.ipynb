```python
# Handling missing values

# Filling missing values with 0 (if appropriate for numerical data)
df.fillna(0, inplace=True)

# Drop rows with missing values (use if missing values are sparse)

df.dropna(inplace=True)

#  Filling missing 'Calories' with the column mean
df['calories'].fillna(df['calories'].mean(), inplace=True)

# Verifying no missing values remain
print("\nData Info After Handling Missing Values:")
print(df.info())
```

```python
# Removing duplicates if any

df.drop_duplicates(inplace=True)

# Cleaning column names (removing spaces, making lowercase, etc.)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Verifying cleaned data

print("\nCleaned Data Sample:")
print(df.head())
```

```python
# Extracting calorie data for analysis
calorie_data = df['calories']

# Summary statistics
calorie_summary = calorie_data.describe()
print(calorie_summary)

# Plotting the distribution of calorie counts
plt.figure(figsize=(10, 6))
sns.histplot(calorie_data, kde=True, bins=20, color='skyblue')
plt.title('Distribution of Calorie Counts Across Menu Items', fontsize=16)
plt.xlabel('Calories', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Analyze trends and patterns in the dataset

# Set plot style
sns.set(style='whitegrid')

# Group data by category and calculate mean nutritional values
category_means = df.groupby('category')[['calories', 'protein','total_fat', 'carbohydrates']].mean()

# Plot the mean nutritional values by category
category_means.plot(kind='bar', figsize=(12, 6), title='Average Nutritional Content by Category')
plt.ylabel('Average Value')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the calculated means for reference
print(category_means)
```

```python
# Setting up the visualizations for calorie distribution and nutritional content

# Bar chart for average calories per category
plt.figure(figsize=(10, 6))
df.groupby('category')['calories'].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Calories per Category')
plt.ylabel('Average Calories')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Histogram for calorie distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['calories'], bins=20, kde=True, color='orange')
plt.title('Calorie Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Box plot for calories by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='calories', data=df, palette='Set2')
plt.title('Calories by Category')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#To compare nutritional characteristics across food categories, 
#I will calculate and visualize the average values of key nutritional metrics (e.g., calories, fat, protein) 
#for each category using grouped bar charts and summary tables.

# Comparing nutritional characteristics across food categories

# Selecting key nutritional columns for comparison
nutritional_columns = ['calories', 'total_fat', 'saturated_fat', 'cholesterol', 'sodium', 'carbohydrates', 'protein']

# Grouping by category and calculating the mean for each nutritional column
category_nutrition = df.groupby('category')[nutritional_columns].mean()

# Displaying the summary table
print(category_nutrition)

# Plotting grouped bar charts for key nutritional metrics
category_nutrition.plot(kind='bar', figsize=(14, 8), colormap='viridis')
plt.title('Average Nutritional Characteristics by Food Category')
plt.ylabel('Average Value')
plt.xlabel('Food Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Creating a heatmap for better visualization of nutritional differences

plt.figure(figsize=(12, 8))
sns.heatmap(category_nutrition, annot=True, fmt='.1f', cmap='YlOrRd', center=0)
plt.title('Nutritional Content Heatmap by Food Category')
plt.ylabel('Food Category')
plt.xlabel('Nutritional Components')
plt.tight_layout()
plt.show()

# Box plots for protein and fat distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.boxplot(x='category', y='protein', data=df, ax=ax1)
ax1.set_title('Protein Content Distribution by Category')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

sns.boxplot(x='category', y='total_fat', data=df, ax=ax2)
ax2.set_title('Total Fat Distribution by Category')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Sort items by calories (both highest and lowest)
# Get top 10 highest calorie items
highest_calories = df.nlargest(10, 'calories')[['category', 'item', 'calories']].reset_index(drop=True)
print("Top 10 Highest Calorie Items:")
print(highest_calories)

print("\
" + "="*50 + "\
")

# Get top 10 lowest calorie items (excluding zero-calorie items)
lowest_calories = df[df['calories'] > 0].nsmallest(10, 'calories')[['category', 'item', 'calories']].reset_index(drop=True)
print("Top 10 Lowest Calorie Items:")
print(lowest_calories)

# Create a visual comparison
plt.figure(figsize=(12, 6))
colors = ['red' if x <= 4 else 'green' for x in range(10)]

# Plot highest calories
plt.subplot(1, 2, 1)
bars = plt.barh(highest_calories['item'].str[:20], highest_calories['calories'], color='red')
plt.title('Highest Calorie Items')
plt.xlabel('Calories')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f'{int(width)}', 
             ha='left', va='center', fontweight='bold')

# Plot lowest calories
plt.subplot(1, 2, 2)
bars = plt.barh(lowest_calories['item'].str[:20], lowest_calories['calories'], color='green')
plt.title('Lowest Calorie Items')
plt.xlabel('Calories')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f'{int(width)}', 
             ha='left', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Calculate average nutritional content by category
nutritional_cols = ['calories', 'total_fat', 'saturated_fat', 'cholesterol', 
                   'sodium', 'carbohydrates', 'dietary_fiber', 'sugars', 'protein']

# Calculate means by category
avg_nutrition = df.groupby('category')[nutritional_cols].mean().round(1)

# Sort by calories to get a sense of the "heaviest" categories
avg_nutrition_sorted = avg_nutrition.sort_values('calories', ascending=False)

print("Average Nutritional Content by Category:")
print(avg_nutrition_sorted)

# Create a more detailed visualization
plt.figure(figsize=(15, 10))

# Create a normalized heatmap to better show the relative values
# Normalize the data
normalized_data = (avg_nutrition_sorted - avg_nutrition_sorted.min()) / (avg_nutrition_sorted.max() - avg_nutrition_sorted.min())

# Create heatmap
sns.heatmap(normalized_data, 
            annot=avg_nutrition_sorted.round(1), 
            fmt='.1f',
            cmap='YlOrRd',
            cbar_kws={'label': 'Normalized Values'})

plt.title('Average Nutritional Content by Category (with actual values)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create radar charts for top 4 categories by calories
top_categories = avg_nutrition_sorted.head(4).index

# Prepare data for radar chart
categories = ['calories/100', 'fat', 'carbs', 'protein', 'sodium/100']

fig, axes = plt.subplots(2, 2, figsize=(15, 15), subplot_kw=dict(projection='polar'))
axes = axes.flatten()

for idx, category in enumerate(top_categories):
    values = [
        avg_nutrition_sorted.loc[category, 'calories']/100,
        avg_nutrition_sorted.loc[category, 'total_fat'],
        avg_nutrition_sorted.loc[category, 'carbohydrates'],
        avg_nutrition_sorted.loc[category, 'protein'],
        avg_nutrition_sorted.loc[category, 'sodium']/100
    ]
    
    # Close the plot by appending the first value
    values += values[:1]
    angles = np.linspace(0, 2*np.pi, len(categories))
    angles = np.concatenate((angles, [angles[0]]))  # complete the circle
    
    ax = axes[idx]
    ax.plot(angles, values)
    ax.fill(angles, values, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    ax.set_title(category)

plt.tight_layout()
plt.show()
```
